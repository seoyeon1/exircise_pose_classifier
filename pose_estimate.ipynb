# Import Dependencies

```python
import pafy
import cv2
import mediapipe as mp
import numpy as np
mp_drawing = mp.solutions.drawing_utils #그리기 유틸리티(점, 밑줄)
mp_pose = mp.solutions.pose #포즈 추정 모델
```

```python
#VIDEO FEED
cap = cv2.VideoCapture() #웹캠 연결
while cap.isOpened(): #웹캠 엑세스 
    ret,frame = cap.read() #ret-> 정적 변수, frame-> 이미지
    cv2.imshow("Mediapipe Feed", frame) #화면에 팝업 제공(이미지 시각화)
    
    if cv2.waitKey(10) & 0xFF == ord('q'): #화면 벗어날 때
        break
        
cap.release() #웹캠 풀기
cv2.destroyAllwindows() #video 피드 닫기
```

# 1. Make Detections

```python
url = "https://www.youtube.com/watch?v=4hYK8ZiDszQ"
video = pafy.new(url)
best = video.getbest(preftype="mp4")

cap = cv2.VideoCapture(best.url)
#cap = cv2.VideoCapture(0) 
## Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
#최소 감지, 추적 신뢰도 50%    
    while cap.isOpened(): 
        ret,frame = cap.read()
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) #bgr->rgb 재정렬
        image.flags.writeable = False #image쓰기가 가능한지 여부 설정 성능 조정 
    
        # Make detection
        results = pose.process(image) #포즈 감지
    
        # Recolor back to BGR
        image.flags.writeable = True #그림을 그릴 수 있는 상태
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 ) #특정 랜드마크 사양 편집
    
        cv2.imshow("Mediapipe Feed", image)
    
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
        
cap.release()
cv2.destroyAllWindows()
```

# 3. Calculate Angles

```python
def calculate_angle(a,b,c):
    a = np.array(a) #First
    b = np.array(b) #Mid
    c = np.array(c) #End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0]) # 1-> y, 0 -> x
    angle = np.abs(radians*180.0/np.pi)
    
    if angle > 180.0: #0-180사이 각도로 변환 - 팔의 최대 각도는 180도 이기 때문
        angle = 360-angle
        
    return angle
```

# 4. Curl Counter

```python
# url = "https://www.youtube.com/watch?v=URVOrAxBDq0"
# video = pafy.new(url)
# best = video.getbest(preftype="mp4")

cap = cv2.VideoCapture(0)

# Curl counter variables
counter = 0
stage = None # curl이 아래쪽 부분인지 위쪽 부분에 있는지 여부를 나타냄

## Setup mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
#최소 감지, 추적 신뢰도 50%    
    while cap.isOpened(): 
        ret,frame = cap.read()
        
        # Recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) #bgr->rgb 재정렬
        image.flags.writeable = False #image쓰기가 가능한지 여부 설정 성능 조정 
    
        # Make detection
        results = pose.process(image) #포즈 감지
    
        # Recolor back to BGR
        image.flags.writeable = True #그림을 그릴 수 있는 상태
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            
            # Get coordinates(좌표 가져옴)
            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
                
            # Calculate angle(계산)
            angle = calculate_angle(shoulder, elbow, wrist)
                
            # Visualize angle(640,480)-> 웹캠 사이즈
            cv2.putText(image, str(angle),
                        tuple(np.multiply(elbow, [640, 480]).astype(int)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255),2,cv2.LINE_AAA
                         )
            
            # Curl counter logic
            if angle> 160: #curl이 아래쪽에 위치하는지 여부
                stage = "down"
            if angle < 30 and stage == "down":
                stage = "up"
                counter += 1
                print(counter)
                
            
            print(landmarks)
        except:
            pass #감지가 없거나 오류가 생길 때 루프를 중단하지 않고 패스
        
        # Render curl counter
        # Setup status box
        cv2.rectangle(image,(0,0),(225,73),(245,117,16),-1)#rectangle 시작점, 끝점, 색상, 선너비
        
        # Rep data
        cv2.putText(image, 'REPS', (15,12),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter), 
                   (10,60),
                   cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)
        
        # Render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 ) #특정 랜드마크 사양 편집
    
        cv2.imshow("Mediapipe Feed", image)
    
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
        
cap.release()
cv2.destroyAllWindows()
```
